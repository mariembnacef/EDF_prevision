```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
```

```python
# 1. Séparer les données
df_train = df[df["Prévision J"] > 0].copy()
df_zero = df[df["Prévision J"] == 0].copy()
# 2. Variables utilisées pour prédire Prévision J
features = ["Prévision J-1", "Heures", "Jour", "Type de jour TEMPO", "Saison", "Weekend"]

# 3. Encodage des variables catégorielles
X_train = pd.get_dummies(df_train[features])
y_train = df_train["Prévision J"]

X_zero = pd.get_dummies(df_zero[features])
X_zero = X_zero.reindex(columns=X_train.columns, fill_value=0)
```

```python
# 4. Entraînement du modèle de régression
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# 5. Prédiction des valeurs manquantes
df_zero["Prévision J"] = model.predict(X_zero)

# 6. Réunir les deux sous-ensembles
df_corrige = pd.concat([df_train, df_zero]).sort_index()
```

```python
df_zero_corrige = df_corrige.loc[df["Prévision J"] == 0]
```
