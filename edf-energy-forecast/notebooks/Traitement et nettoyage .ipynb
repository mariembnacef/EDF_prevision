```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
```

```python
# 1. S√©parer les donn√©es
df_train = df[df["Pr√©vision J"] > 0].copy()
df_zero = df[df["Pr√©vision J"] == 0].copy()
# 2. Variables utilis√©es pour pr√©dire Pr√©vision J
features = ["Pr√©vision J-1", "Heures", "Jour", "Type de jour TEMPO", "Saison", "Weekend"]

# 3. Encodage des variables cat√©gorielles
X_train = pd.get_dummies(df_train[features])
y_train = df_train["Pr√©vision J"]

X_zero = pd.get_dummies(df_zero[features])
X_zero = X_zero.reindex(columns=X_train.columns, fill_value=0)
```

```python
# 4. Entra√Ænement du mod√®le de r√©gression
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# 5. Pr√©diction des valeurs manquantes
df_zero["Pr√©vision J"] = model.predict(X_zero)

# 6. R√©unir les deux sous-ensembles
df_corrige = pd.concat([df_train, df_zero]).sort_index()
```

```python
df_zero_corrige = df_corrige.loc[df["Pr√©vision J"] == 0]
```
